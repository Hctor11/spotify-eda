```python
import pandas as pd
import numpy as numpy
import seaborn as sns 
import matplotlib.pyplot as plt

sns.set_style("darkgrid")
```

```python
# load data
df = pd.read_csv("./data/data.csv")
df.drop("Unnamed: 0", axis=1, inplace=True)
df
```

```python
# Data analysis
# Top 5 most popular artist

top_five_artist = df.groupby("artist").count().sort_values(by="song_title", ascending=False)["song_title"][:5]
top_five_artist
```

```python
top_five_artist.plot.barh()
plt.show()
```

```python
# Top 5 loudest tracks
top_five_loud = df[["loudness","song_title"]].sort_values(by="loudness", ascending=True)[:5]
top_five_loud
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x="loudness", y="song_title", data=top_five_loud)
plt.title("top 5 loudest tracks")
plt.show()
```

```python
# Top 5 danceable songs
top_danceable_songs = df[["danceability", "song_title", "artist"]].sort_values(by="danceability", ascending=False)[:5]
top_danceable_songs
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(x="danceability", y="artist", data=top_danceable_songs, palette="muted")
plt.show()
```

```python
# Top 10 instrumental tracks
top_ten_instrumental = df[["instrumentalness", "song_title", "artist"]].sort_values(by="instrumentalness", ascending=False)[:5]
top_ten_instrumental
```

```python
plt.figure(figsize=(12,7))
plt.pie(x="instrumentalness", data=top_ten_instrumental, labels=top_ten_instrumental.song_title)
```
